```python
from tkinter import *
from tkinter import filedialog
import pandas as pd

sc = Tk()
sc.title("Danh sách Đăng ký")
sc.geometry("500x300")

CheckVar1 = IntVar()
CheckVar2 = IntVar()
CheckVar3 = IntVar()

url = filedialog.askopenfilename(filetypes = (("Excel files","*.xlsx"),("all files","*.*")))
excel = pd.read_excel(url)

tit_bar = list(excel.columns.values)
D = {}
for i in tit_bar:
       D[i] = list(excel[i])
Last_name = []
First_name = []
for i in D["Họ và tên"]:
    t = i.rfind(" ")
    a = i[:t]
    b = i[t+1:]
    Last_name.append(a)
    First_name.append(b)
DOB = []
for i in range(len(D["Ngày sinh"])):
    i = str(D["Ngày sinh"][i])+"-"+str(D["Tháng sinh"][i])+"-"+str(D["Năm sinh"][i])
    DOB.append(i)
Exam = []
for i in range(len(D["Môn thi"])):
    if D["Môn thi"][i] == "W" or D["Môn thi"][i] == "E" or D["Môn thi"][i] == "PP" :
        i = D["Môn thi"][i] + str(D["Phiên bản"][i])[-2:]
    else:
        i = D["Môn thi"][i] + "-" + D["Phiên bản"][i]
    Exam.append(i)
SM = list(D["STT"])
STT = list(SM)
ID = list(D["Số CMND"])
Username = list(ID)
Note = list(STT)
    
L_cot = []
D_data = {}
def data_input():
    caccot = list(excel.columns.values)
    for child in content.winfo_children():
        child.destroy()
    for i in range(2,len(caccot)-1):
        L_cot.append(Label(content,text = caccot[i],font = ("Arial",10),borderwidth=5, relief="groove"))
        L_cot[-1].grid(column = i-1,row = 1,sticky = "we")
        D_data[caccot[i]] = list(excel[caccot[i]])
    for i in range(15):
        dem = 1
        for key in D_data:
            D_data[key][i] = Label(content,text = D_data[key][i],font = ("Arial",10),borderwidth=5, relief="groove",justify = "left")
            D_data[key][i].grid(column = dem, row = i+2,sticky = "we")
            dem += 1

def open_win():
    new = Toplevel(sc)
    new.geometry("1100x600")
    new.title("Danh sách Dự thi")
    def hide_info():
        if CheckVar1.get() == 1:
            for i in range(len(excel)):
                if D["Ca thi"][i] == "Ca 1":
                    SM[i].grid(column = 1, row = i+7, sticky = "E")
                    STT[i].grid(column = 2, row = i+7,sticky = "we")
                    Last_name[i].grid(column = 3, row = i+7,sticky = "we")
                    First_name[i].grid(column = 4, row = i+7,sticky = "we")
                    DOB[i].grid(column = 5, row = i+7,sticky = "we")
                    ID[i].grid(column = 6, row = i+7,sticky = "we")
                    Username[i].grid(column = 7, row = i+7,sticky = "we")
                    Exam[i].grid(column = 8, row = i+7,sticky = "we")
                    Note[i].grid(column = 11, row = i+7,sticky = "we")
        else:
            for i in range(len(excel)):
                if D["Ca thi"][i] == "Ca 1": 
                    SM[i].grid_forget()
                    STT[i].grid_forget()                       
                    Last_name[i].grid_forget()
                    First_name[i].grid_forget()
                    DOB[i].grid_forget()
                    ID[i].grid_forget()
                    Username[i].grid_forget()
                    Exam[i].grid_forget()
                    Note[i].grid_forget()
        if CheckVar2.get() == 1:
            for i in range(len(excel)):
                if D["Ca thi"][i] == "Ca 2":
                    SM[i].grid(column = 1, row = i+7, sticky = "E")
                    STT[i].grid(column = 2, row = i+7,sticky = "we")
                    Last_name[i].grid(column = 3, row = i+7,sticky = "we")
                    First_name[i].grid(column = 4, row = i+7,sticky = "we")
                    DOB[i].grid(column = 5, row = i+7,sticky = "we")
                    ID[i].grid(column = 6, row = i+7,sticky = "we")
                    Username[i].grid(column = 7, row = i+7,sticky = "we")
                    Exam[i].grid(column = 9, row = i+7,sticky = "we")
                    Note[i].grid(column = 11, row = i+7,sticky = "we")                         
        else:
            for i in range(len(excel)):
                if D["Ca thi"][i] == "Ca 2": 
                    SM[i].grid_forget()
                    STT[i].grid_forget()                       
                    Last_name[i].grid_forget()
                    First_name[i].grid_forget()
                    DOB[i].grid_forget()
                    ID[i].grid_forget()
                    Username[i].grid_forget()
                    Exam[i].grid_forget()
                    Note[i].grid_forget()
        if CheckVar3.get() == 1:
            for i in range(len(excel)):
                if D["Ca thi"][i] == "Ca 3":
                    SM[i].grid(column = 1, row = i+7, sticky = "E")
                    STT[i].grid(column = 2, row = i+7,sticky = "we")
                    Last_name[i].grid(column = 3, row = i+7,sticky = "we")
                    First_name[i].grid(column = 4, row = i+7,sticky = "we")
                    DOB[i].grid(column = 5, row = i+7,sticky = "we")
                    ID[i].grid(column = 6, row = i+7,sticky = "we")
                    Username[i].grid(column = 7, row = i+7,sticky = "we")
                    Exam[i].grid(column = 10, row = i+7,sticky = "we")
                    Note[i].grid(column = 11, row = i+7,sticky = "we")                          
        else:
            for i in range(len(excel)):
                if D["Ca thi"][i] == "Ca 3": 
                    SM[i].grid_forget()
                    STT[i].grid_forget()                       
                    Last_name[i].grid_forget()
                    First_name[i].grid_forget()
                    DOB[i].grid_forget()
                    ID[i].grid_forget()
                    Username[i].grid_forget()
                    Exam[i].grid_forget()
                    Note[i].grid_forget()
                                       
    logo = Label(new, text="IIG VIETNAM", font=("Arial",14,"bold"), fg="blue").grid(column=1, row=1, columnspan=2, padx=(30,0), pady=(0,20))
    tit = Label(new, text="ATTENDANCE LIST", font=("Arial",18,"bold")).grid(column=2, row=2, columnspan=11, pady=(0,20))
    room = Label(new, text = "Phòng thi", font=("Arial",10)).grid(column=2, row=3, sticky = W)
    test_date = Label(new, text = "Ngày thi", font=("Arial",10)).grid(column=2, row=4, sticky = W, pady=(0,20))
    et_room = Entry(new, font=("Arial",10), width=15).grid(column=3, row=3, columnspan=2, sticky = W)
    et_test_date = Entry(new, font=("Arial",10), width=15).grid(column=3, columnspan=2, row=4, sticky = W, pady=(0,20))
    lb1 = Label(new, text="Số máy",font=("Arial",10,"bold"), width=7, height=2, relief="solid", borderwidth=1).grid(row=5, column=1, padx=(30,0))
    lb2 = Label(new, text="STT",font=("Arial",10,"bold"), width=8, height=2, relief="solid", borderwidth=1).grid(row=5, column=2)
    lb3 = Label(new, text="Họ và đệm",font=("Arial",10,"bold"), width=20, height=2, relief="solid", borderwidth=1).grid(row=5, column=3)
    lb4 = Label(new, text="Tên",font=("Arial",10,"bold"), width=12, height=2, relief="solid", borderwidth=1).grid(row=5, column=4)
    lb5 = Label(new, text="Ngày sinh",font=("Arial",10,"bold"), width=15, height=2, relief="solid", borderwidth=1).grid(row=5, column=5)
    lb6 = Label(new, text="CMND/HC",font=("Arial",10,"bold"), width=15, height=2, relief="solid", borderwidth=1).grid(row=5, column=6)
    lb7 = Label(new, text="User",font=("Arial",10,"bold"), width=15, height=2, relief="solid", borderwidth=1).grid(row=5, column=7)
    lb8_1 = Checkbutton(new, command = hide_info, text="C1", variable = CheckVar1, onvalue = 1, offvalue = 0, font=("Arial",10,"bold"), width=4, height=2, relief="solid", borderwidth=1)
    lb8_2 = Checkbutton(new, command = hide_info, text="C2", variable = CheckVar2, onvalue = 1, offvalue = 0, font=("Arial",10,"bold"), width=4, height=2, relief="solid", borderwidth=1)
    lb8_3 = Checkbutton(new, command = hide_info, text="C3", variable = CheckVar3, onvalue = 1, offvalue = 0, font=("Arial",10,"bold"), width=4, height=2, relief="solid", borderwidth=1)       
    lb8_1.grid(row=5, column=8)
    lb8_2.grid(row=5, column=9)
    lb8_3.grid(row=5, column=10)
    lb9 = Label(new, text="Ghi chú",font=("Arial",10,"bold"), width=11, height=2, relief="solid", borderwidth=1)
    lb9.grid(row=5, column=11, padx=(0,30))
    for i in range(len(excel)):
        SM[i] = Entry(new, relief="groove", width =5)
        STT[i] = Label(new, text = i+1, relief="groove")   
        Last_name[i] = Label(new, text = Last_name[i], relief="groove")
        First_name[i] = Label(new, text = First_name[i], relief="groove")
        DOB[i] = Label(new, text = DOB[i], relief="groove")
        ID[i] = Label(new, text = ID[i], relief="groove")
        Username[i] = Label(new, text = Username[i], relief="groove")
        Exam[i] = Label(new, text = Exam[i], relief="groove")
        Note[i] = Entry(new, relief="groove")

def step1():
    bt2.pack()
def step2():
    bt3.pack()
def step3():
    bt4.pack()     
    bt5.pack()

bt1 = Button(sc, text = "BRAND NEW DAY!", command = step1).pack()
bt2 = Button(sc, text = "LET 'S WORK!", command = step2)
bt3 = Button(sc, text = "THIS JOB IS YOUR BREAD AND BUTTER!", command = step3)
bt4 = Button(sc,text = 'DATA INPUT' ,width = 20, command = data_input)   
bt5 = Button(sc, text= "DATA OUTPUT", width = 20, command = open_win)
content = Frame(sc)
content.pack()

sc.mainloop()
```
